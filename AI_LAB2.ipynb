```python
import random
class Environment:
  def __init__(self):
    self.servers=[random.choice(["Underloaded","Balanced","Overloaded"]) for _ in range(5)]
  def get_percept(self,position):
    return self.servers[position]
  def balance(self,position):
    self.servers[position]="Balanced"
  def display(self,position):
    servers_agent=self.servers[:]
    servers_agent[position]+="游놓"
    print(servers_agent)

class Agent:
  def __init__(self):
    self.position=0
    self.over=[]
    self.under=[]
  def scan(self,position,percept):
    if percept=="Overloaded":
      self.over.append(position)
    if percept=="Underloaded":
      self.under.append(position)
  def move(self):
    if self.position<=4:
      self.position+=1
    return self.position

def runAgent(agent, environment):
  print("Initial System Check:")
  environment.display(0)

  while agent.position <= 4:
    percept = environment.get_percept(agent.position)
    agent.scan( agent.position,percept)
    agent.move()
  #agent.position=0
  print("\nBalancing Servers:")
  if len(agent.under)==0:
    print("No underloaded servers")
  elif len(agent.over)==0:
    print("No overloaded servers")
  else:
    print("Moving Tasks from Servers")
    for position in agent.over:
      environment.balance(position)
      print(position,", ")
    print("To Servers")
    for position in agent.under:
      print(position,", ")
      environment.balance(position)



  print("\nFinal Servers Check:")
  environment.display(0)

agent = Agent()
environment = Environment()
runAgent(agent, environment)




```

```python
import random

class Environment:
    def __init__(self):
        self.tasks = [random.choice(["Completed", "Failed"]) for _ in range(5)]

    def get_task_status(self, position):
        return self.tasks[position]

    def retry_task(self, position):
        self.tasks[position] = "Completed"  # Assume retry is always successful

    def display_tasks(self, agent_position=None):
        print("\nCurrent Backup Tasks Status:")
        if agent_position is not None:
            tasks_with_agent = self.tasks[:]
            tasks_with_agent[agent_position] = tasks_with_agent[agent_position] + " 游뱄"  # Indicate agent's position
            print(tasks_with_agent)
        else:
            print(self.tasks)

class BackupManagementAgent:
    def __init__(self):
        self.position = 0
        self.failed_tasks = []

    def scan(self, position, status):
        if status == "Failed":
            self.failed_tasks.append(position)

    def move(self):
        if self.position < len(self.failed_tasks) -1 :  # Change here, move to the next failed task if present
          self.position += 1
          return self.position


def runAgent(agent, environment):
    print("Initial Backup Check:")
    environment.display_tasks()

    for i in range(len(environment.tasks)):
        status = environment.get_task_status(i)
        agent.scan(i, status)

    print("\nRetrying Failed Backups:")
    for position in agent.failed_tasks:
        environment.retry_task(position)
        print(f"Task {position} retried and completed.")

    print("\nFinal Backup Check:")
    environment.display_tasks()

agent = BackupManagementAgent()
environment = Environment()
runAgent(agent, environment)
```

```python
import random

class Environment:
    def __init__(self):
        self.components = [random.choice(["Safe", "Low Risk", "High Risk"]) for _ in range(9)]

    def get_percept(self, position):
        return self.components[position]

    def patch(self, position):
        if self.components[position] == "Low Risk":
            self.components[position] = "Safe"

    def display(self, agent_position=None):
        components_agent = self.components[:]
        if agent_position is not None:
            components_agent[agent_position] += " 游놓"
        print(components_agent)

class UtilityBasedAgent:
    def __init__(self):
        self.position = 0
        self.low_risk = []
        self.high_risk = []

    def scan(self, position, percept):
        if percept == "Low Risk":
            print(f"Low Risk Vulnerability Detected: ", position)
            self.low_risk.append(position)
        elif percept == "High Risk":
            print(f" High Risk Vulnerability Detected: " ,position)
            self.high_risk.append(position)

    def move(self):
        if self.position < 8:
            self.position += 1
        return self.position

def runAgent(agent, environment):
    print(" Initial System Check:")
    environment.display()

    while agent.position <= 8:
        percept = environment.get_percept(agent.position)
        agent.scan(agent.position, percept)
        agent.position+=1
    agent.position=0
    print("\n Patching Low Risk Vulnerabilities:")
    if not agent.low_risk:
        print(" No Low Risk vulnerabilities to patch.")
    else:
        for position in agent.low_risk:
            environment.patch(position)
            print("Patched Component: " ,position)

    print("\n High Risk Vulnerabilities Remain Unpatched (Requires Premium Service):")
    if not agent.high_risk:
        print(" No High Risk vulnerabilities detected.")
    else:
        for position in agent.high_risk:
            print(f" Component:" , position ,"still vulnerable!")

    print("\n Final System Check:")
    environment.display()

agent = UtilityBasedAgent()
environment = Environment()
runAgent(agent, environment)
```

```python
import random

class Environment:
    def __init__(self):
        self.locations = ["Corridor", "Medicine Storage", "Patient Room 1", "Patient Room 2", "Nurse Station"]
        self.medicine_schedule = {
            "Patient Room 1": {"medicine": "MedA", "time": "9 AM"},
            "Patient Room 2": {"medicine": "MedB", "time": "10 AM"}
        }
        self.current_time = "9 AM"

    def get_percept(self, position):
        location = self.locations[position]
        if location in self.medicine_schedule and self.medicine_schedule[location]['time'] == self.current_time:
            return f"Deliver {self.medicine_schedule[location]['medicine']} to {location}"
        elif location == "Medicine Storage":
            return "Pick up medicine"
        elif location == "Nurse Station":
            return "Alert staff if needed"
        else:
            return f"Moving through {location}"

    def display(self, position):
        hospital_layout = self.locations[:]
        hospital_layout[position] += " 游띶"
        print(hospital_layout)


class Agent:
    def __init__(self):
        self.position = 0  # Start at Corridor
        self.carrying = None

    def move(self, direction, env):
        if direction == "forward" and self.position < len(env.locations) - 1:
            self.position += 1
        elif direction == "backward" and self.position > 0:
            self.position -= 1
        print(f"Moved to {env.locations[self.position]}")

    def act(self, percept, env):
        if "Pick up medicine" in percept:
            self.carrying = "MedA" if env.current_time == "9 AM" else "MedB"
            print(f"Picked up {self.carrying}")
        elif "Deliver" in percept and self.carrying:
            print(f"Scanning patient ID at {env.locations[self.position]}")
            print(f"Delivered {self.carrying} to {env.locations[self.position]}")
            self.carrying = None
        elif "Alert staff" in percept:
            print("Alerting staff at Nurse Station for assistance.")
        else:
            print(percept)



env = Environment()
robot = Agent()

for _ in range(10):
    env.display(robot.position)
    percept = env.get_percept(robot.position)
    robot.act(percept, env)
    robot.move("forward", env)
```
